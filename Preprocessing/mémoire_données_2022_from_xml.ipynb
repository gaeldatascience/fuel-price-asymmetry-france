```python
import xml.etree.ElementTree as ET
import pandas as pd

# Charger le fichier XML
file_path = "C:/Users/zetru/Downloads/PrixCarburants_annuel_2022/PrixCarburants_annuel_2022.xml"
tree = ET.parse(file_path)
root = tree.getroot()

# Initialiser une liste pour stocker les données
data = []

# Parcourir tous les éléments 'pdv' dans le fichier XML
for pdv in root.findall('pdv'):
    # Informations de base du point de vente
    base_info = {
        'id': pdv.get('id'),
        'type' : pdv.get('pop'),
        'latitude': pdv.get('latitude'),
        'longitude': pdv.get('longitude'),
        'code_postal': pdv.get('cp'),
        'ville': pdv.find('ville').text if pdv.find('ville') is not None else None,
    }
    
    # Extraire les informations de prix pour chaque carburant disponible
    for prix in pdv.findall('prix'):
        # Créer une nouvelle ligne pour chaque type de carburant avec ses informations de prix
        prix_data = {
            **base_info,  # Inclure les informations de base du pdv
            'nom_carburant': prix.get('nom'),
            'id_carburant': prix.get('id'),
            'maj': prix.get('maj'),
            'valeur': prix.get('valeur')
        }
        # Ajouter la ligne au dataset
        data.append(prix_data)

# Créer un DataFrame à partir de la liste de données
df = pd.DataFrame(data)
df['valeur'] = df['valeur'].astype(float)
# Afficher les premières lignes du DataFrame pour vérification
print(df.head())
```
