**Purpose of the notebook: Drop stations selling both SP95 (old) & E10 (used for the study).**

```python
import os
os.chdir("C:/Users/zetru/OneDrive/Mémoire python/")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 16, 9
plt.rcParams['axes.grid'] = True
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
df=pd.read_hdf('df_v5.h5','df_v5')
df
```

```python
stations_sp95 = df[df['nom_carburant'] == 'SP95']['id'].unique()
stations_e10 = df[df['nom_carburant'] == 'E10']['id'].unique()
stations_both = list(set(stations_sp95) & set(stations_e10))
df = df[(df['id'].isin(stations_both) & (df['nom_carburant'] != 'SP95')) | (~df['id'].isin(stations_both))]
```

```python
df
```

```python
df['nom_carburant'] = df['nom_carburant'].replace(['SP95', 'E10'], 'SP95E5_E10')
```
