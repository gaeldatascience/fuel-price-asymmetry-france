```python
df_last=df.groupby(['cp','id', 'nom_carburant', 'maj']).last().reset_index()
```

```python
# Ici, je filtre les valeurs extrêmes de mon df :
```

```python
carburants_concernes = ['Gazole', 'SP95', 'SP98', 'E10']

df_last=df_last[~((df_last['nom_carburant'].isin(carburants_concernes)) & (df_last['valeur'] < 1.3))]
```

```python
df_last=df_last[~((df_last['valeur'] < 0.3))].reset_index(drop=True)
```

```python
df_last=df_last[df_last['valeur']!=1].reset_index(drop=True)
```

```python
df_last=df_last[~((df_last['valeur'] >2.78))].reset_index(drop=True)
```

```python
GPL_E85 = ['GPLc', 'E85']

df_last=df_last[~((df_last['nom_carburant'].isin(GPL_E85)) & (df_last['valeur'] >1.4))].reset_index(drop=True)
```

```python
for carburant in df_last['nom_carburant'].unique():
    print(df_last[df_last['nom_carburant']==carburant][['id','nom_carburant','valeur','maj']].nlargest(150,'valeur'))
```
