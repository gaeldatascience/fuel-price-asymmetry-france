```python
df_signi=pd.DataFrame(df.groupby(["nom_carburant","maj","type"])['valeur'].mean()).reset_index()
```

```python
df_signi
```

```python
df_signi_a=df_signi[(df_signi['type']=='A')]
```

```python
df_signi_r=df_signi[(df_signi['type']=='R')]
```

```python
import numpy as np
from scipy import stats
resultat=[]

# Effectuer le test de Student
for carburant in df_signi_a['nom_carburant'].unique():
    df_signi_atest=df_signi_a[df_signi_a['nom_carburant']==carburant].iloc[:,-1]
    df_signi_rtest=df_signi_r[df_signi_r['nom_carburant']==carburant].iloc[:,-1]
    t_stat, p_value = stats.ttest_ind(df_signi_atest, df_signi_rtest, equal_var=True)
    if p_value < 0.05:
        interprétation="Moyennes significativement différentes."
    else:
        interprétation="Pas de différence significative."
    
    resultat.append({
        'carburant':carburant,
        'Statistique T':t_stat,
        'P-value':p_value,
        'Interprétation':interprétation})
print('Test de Student :')
pd.DataFrame(resultat)
```

```python
pd.DataFrame(resultat).to_latex('tex/Test de Student.tex',index=False)
```

```python
resultat2=[]
for carburant in df_signi_a['nom_carburant'].unique():
    df_signi_atest=df_signi_a[df_signi_a['nom_carburant']==carburant].iloc[:,-1]
    df_signi_rtest=df_signi_r[df_signi_r['nom_carburant']==carburant].iloc[:,-1]
    w_stat, p_value = stats.levene(df_signi_atest, df_signi_rtest)
    # Interprétation du résultat
    if p_value > 0.05:
        interprétation='Variances égales entre les groupes.'
    else:
        interprétation='Différence significative dans les variances.'
    resultat2.append({
        'carburant':carburant,
        'Statistique W':w_stat,
        'P-value':p_value,
        'Interprétation':interprétation})
print('Test de Levene :')
pd.DataFrame(resultat2)
```

```python
pd.DataFrame(resultat2).to_latex('tex/Test de Levene.tex',index=False)
```
